Newspaper Libary is used to obtain article texts from the listed URLs

```python
def getArticleText(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        text=article.text
    except:
        print("****failed to obtain text******",article.url)
        text=""
    return text
```

```python
newsLabels["article_text"]=[getArticleText(url) for url in newsLabels["url"]]
```

Articles with negative/positve ratings for both Republican and Democratic Parties are removed.

```python
def findDouble(df):
    if (df['democrat.vote'] == "SomewhatNegative" )or (df['democrat.vote'] == "Negative"):
        if (df['republican.vote'] == "SomewhatNegative" )or (df['republican.vote'] == "Negative"):
            return "Y"
    if (df['democrat.vote'] == "SomewhatPositive" )or (df['democrat.vote'] == "Positive"):
        if (df['republican.vote'] == "SomewhatPositive" )or (df['republican.vote'] == "Positive"):
            return "Y"
    else:
        return "N"
```

```python
NewsLabelsWithNoNa['double_label'] = NewsLabelsWithNoNa.apply (lambda df: findDouble(df), axis=1)
```

```python
NewsLabelsWithNoNa
```

```python
doubletodelete=NewsLabelsWithNoNa[NewsLabelsWithNoNa['double_label']=="Y"].index
```

```python
NewsLabelsWithNoNa.drop(doubletodelete,inplace=True)
```

Articles are cleaned so that only textual content remains for each article.

```python
def clean_articles(txt):
    txt=list(txt)
    txt=[character for character in txt if character.isalnum() or character.isspace()]
    txt="".join(txt)
 
    return txt
```

```python
NewsLabelsWithNoNa["article_text"] = NewsLabelsWithNoNa["article_text"].apply(clean_articles)
```

```python
NewsLabelsWithNoNa.drop(columns=['Unnamed: 0','double_label'],inplace=True)
```

```python
NewsLabelsWithNoNa
```

Articles with Neutral ratings for both Republican and Democrat are considered "Neutral"

```python
#find double neutrals
def findDoubleNeutral(df):
    if (df['democrat.vote'] == "Neutral" ):
        if (df['republican.vote'] == "Neutral" ):
            return "Y"
    if (df['republican.vote'] == "Neutral"):
        if (df['democrat.vote'] == "Neutral" ):
            return "Y"
    else:
        return "N"
```

```python
NewsLabelsWithNoNa['doubleN_label'] = NewsLabelsWithNoNa.apply (lambda df: findDoubleNeutral(df), axis=1)
```

```python
shuffled_df=NewsLabelsWithNoNa.sample(frac=1,random_state=4)

non_neutraldf=shuffled_df[shuffled_df["doubleN_label"]!="Y"]
```

```python
neutral_df=shuffled_df.loc[shuffled_df["doubleN_label"]=="Y"].sample(n=500,random_state=10)
```
